```python
sst = ds['sst']
date = datetime(2025, 8, 7)  # Example: same as Sentinel-2 acquisition
sst_day = sst.sel(valid_time=slice(f"{date:%Y-%m-%d}T00", f"{date:%Y-%m-%d}T23"))
sst_mean = sst_day.mean(dim="valid_time")
print(sst_mean)
```
